```python
# Compute daily percentage changes
RR['pct_change_open'] = RR['open'].pct_change()
RR['pct_change_high'] = RR['high'].pct_change()
RR['pct_change_low'] = RR['low'].pct_change()
RR['pct_change_close'] = RR['close'].pct_change()
RR['pct_change_volume'] = RR['volume'].pct_change()

#RR['pct_change_start'] = RR['start_value'].pct_change()

#RR = RR.dropna()
#RR
```

```python
# Compute log returns
RR['log_return_open'] = np.log(RR['open'] / RR['open'].shift(1))
RR['log_return_high'] = np.log(RR['high'] / RR['high'].shift(1))
RR['log_return_low'] = np.log(RR['low'] / RR['low'].shift(1))
RR['log_return_close'] = np.log(RR['close'] / RR['close'].shift(1))
RR['log_return_volume'] = np.log(RR['volume'] / RR['volume'].shift(1))

#RR['log_return_start'] = np.log(RR['start_value'] / RR['start_value'].shift(1))

RR
```

```python
# Compute RSI
window_length_RSI = 14
delta = RR['close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=window_length_RSI).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=window_length_RSI).mean()
rs = gain / loss
RR['rsi'] = 100 - (100 / (1 + rs))
```

```python
# Rolling standard deviation (volatility)
RR['volatility'] = RR['close'].rolling(window=5).std()

```

```python
# Detrended price
rolling_mean = RR['close'].rolling(window=14).mean()
RR['dpo'] = RR['close'] - rolling_mean.shift(14 // 2)
```

```python
#High/Close and Low/Close ratios:

RR['high_low_ratio'] = RR['high'] / RR['low']
RR['high_close_ratio'] = RR['high'] / RR['close']
RR['low_close_ratio'] = RR['low'] / RR['close']
```

```python
# Short-term and long-term moving averages
RR['short_ma'] = RR['close'].rolling(window=12).mean()
RR['long_ma'] = RR['close'].rolling(window=26).mean()

# MACD
RR['macd'] = RR['short_ma'] - RR['long_ma']
RR['macd_signal'] = RR['macd'].rolling(window=9).mean()
```
